```python
import pandas as pd

# Carregue o seu conjunto de dados original
data = pd.read_csv('Simples_Carga_de_Energia_Dia_Hora_data.csv', sep=',')
data
```

```python
# Renomeie as colunas
data.columns = ['Data', 'int_hour', 'subsistema', 'load_mwh']
data
```

```python
from datetime import datetime

data['Data'] = data['Data'].apply(lambda x: datetime.strptime(x, "%d/%m/%Y %H:%M:%S").strftime("%Y-%m-%d"))
data
```

```python
import pandas as pd
import numpy as np

horas = np.tile(np.arange(24), len(data) // 24 + 1)[:len(data)]
data['int_hour'] = horas
```

```python
data['is_outlier'] = False
```
